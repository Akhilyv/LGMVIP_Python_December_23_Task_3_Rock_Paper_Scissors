```python
from PIL import Image, ImageTk
import IPython.display as display
import random
import time

def generate_computer_move():
    moves = ["Rock", "Paper", "Scissors"]
    return random.choice(moves)

def check_winner(player_move, computer_move):
    if player_move == computer_move:
        return "It's a tie!"
    elif (
        (player_move == "Rock" and computer_move == "Scissors") or
        (player_move == "Paper" and computer_move == "Rock") or
        (player_move == "Scissors" and computer_move == "Paper")
    ):
        return "You win!"
    else:
        return "Computer wins!"

def play_game(user_choice):
    computer_choice = generate_computer_move()
    result = check_winner(user_choice, computer_choice)

    # Display images with full path
    display.display(display.Image(filename=f'/content/sample_data/{user_choice.lower()}.png', format='png', width=100, height=100))
    display.display(display.Image(filename=f'/content/sample_data/{computer_choice.lower()}.png', format='png', width=100, height=100))
    print(result)

# Main loop
while True:
    # Get user input
    user_choice = input("Enter your move (Rock/Paper/Scissors) or 'Quit' to exit: ")

    # Validate user input
    if user_choice.lower() == "quit":
        print("Exiting the game.")
        break
    elif user_choice.lower() not in ["rock", "paper", "scissors"]:
        print("Invalid move. Please enter Rock, Paper, or Scissors.")
    else:
        # Play the game
        print("Game starting in:")
        for i in range(5, 0, -1):
            print(i, end=' ', flush=True)
            time.sleep(1)
        print()
        play_game(user_choice.capitalize())
```
